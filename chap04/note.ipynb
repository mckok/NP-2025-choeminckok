## 1부터45까지의수중에서6개를선택하여로또번호를만드는프로그램

```python
import random
pick = set()
while len(pick) < 6:
    n = random.randint(1,45)
    if n not in pick:
        pick.add(n)
        print(pick)
        print(sorted(pick)) 
```
